*Iris Flower Classification*

```python
# import the libraries
import pandas as pd
import numpy as np
```

```python
# connect the csv file (dataset)
df = pd.read_csv("/content/Iris.csv")
```

```python
# read the dataset
df
```

```python
# return the number of rows and columns in dataset
df.shape
```

```python
# drop the "Id" column
df = df.drop(columns=["Id"])
```

```python
# return the first 5 values (rows) of dataset
df.head()
```

```python
#assign "1:Iris-setosa" "2:Iris-versicolor" "3:Iris-virginica"

df["Species"].replace({"Iris-setosa":1, "Iris-versicolor":2, "Iris-virginica":3},inplace = True)
```

```python
df
```

```python
x=pd.DataFrame(df,columns=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]).values
```

```python
x
```

```python
y = df.Species.values.reshape(-1,1)
```

```python
y
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30, random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
k=6
knclr=KNeighborsClassifier(k)
```

```python
knclr.fit(x_train,y_train)
```

```python
y_pred=knclr.predict(x_test)
```

```python
metrics.accuracy_score(y_test,y_pred)*100
```
